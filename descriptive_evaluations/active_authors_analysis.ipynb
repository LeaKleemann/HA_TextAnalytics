```python
#Forschungsaspekt: Welche besonders aktiven Autoren gibt es?
import pandas as pd
import matplotlib.pyplot as plt
import glob

csv_folder_file_path = r"C:\HLAG\Apps\Self_Learn_Workspace\Python_Natural_Language_processing_NAKPrep\PY Hausarbeit\HA_TextAnalytics\lens_org_data_files\*.csv"
all_csv_file_paths = glob.glob(csv_folder_file_path)

individual_dataframes = [pd.read_csv(file) for file in all_csv_file_paths]
summarized_dataframe = pd.concat(individual_dataframes, ignore_index=True) 

```

```python
columnName = ("Author/s")

counts_per_author = summarized_dataframe[columnName].dropna().astype(str)
counts_per_author = counts_per_author[~counts_per_author.str.contains("null", case=False)]
counts_per_author = counts_per_author.value_counts()

most_active_authors = counts_per_author.nlargest(10)

print(type(most_active_authors))
print(most_active_authors)

# Plot as a bar chart
plt.figure(figsize=(12, 6))
most_active_authors.plot(kind='bar', color='skyblue', edgecolor='black')

# Customize the chart
plt.xlabel("Publisher", fontsize=12)
plt.ylabel("Menge der Verffentlichungen", fontsize=12)
plt.title("Anzahl Verffentlichungen pro Publisher", fontsize=14)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
