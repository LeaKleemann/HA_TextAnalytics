```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

csv_folder_file_path = r"C:\HLAG\Apps\Self_Learn_Workspace\Python_Natural_Language_processing_NAKPrep\PY Hausarbeit\HA_TextAnalytics\lens_org_data_files\*.csv"
all_csv_file_paths = glob.glob(csv_folder_file_path)

individual_dataframes = [pd.read_csv(file) for file in all_csv_file_paths]
summarized_dataframe = pd.concat(individual_dataframes, ignore_index=True)

columnName = ("Source Country")

country_counts = summarized_dataframe[columnName].dropna().astype(str).value_counts()
```

```python
high_volume_country_counts = country_counts.nlargest(30)

print(high_volume_country_counts)
print(len(high_volume_country_counts))

# Plot as a bar chart
plt.figure(figsize=(20, 6))
high_volume_country_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Customize the chart
plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Articles", fontsize=12)
plt.title("Number of Articles per Country", fontsize=14)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
#Im obigen Ergebnis gibt es sehr starke Top 5. Hier sollen noch einmal die Lndern unterhalb der top 5 besser zur Geltung kommen.
#Daher werden die Top 5 (oberhalb 10000) ausgeschlossen und dann das Diagramm wiederholt
lower_volume_country_counts = country_counts[country_counts <= 10000].nlargest(30)
plt.figure(figsize=(20, 6))
lower_volume_country_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Customize the chart
plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Articles", fontsize=12)
plt.title("Number of Articles per Country", fontsize=14)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
