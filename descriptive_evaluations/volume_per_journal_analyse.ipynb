```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

csv_folder_file_path = r"C:\HLAG\Apps\Self_Learn_Workspace\Python_Natural_Language_processing_NAKPrep\PY Hausarbeit\HA_TextAnalytics\lens_org_data_files\*.csv"
all_csv_file_paths = glob.glob(csv_folder_file_path)

individual_dataframes = [pd.read_csv(file) for file in all_csv_file_paths]
summarized_dataframe = pd.concat(individual_dataframes, ignore_index=True)

columnName = ("Publisher")

counts_per_publisher = summarized_dataframe[columnName].dropna().astype(str).value_counts()
```

```python
high_volume_publishers = counts_per_publisher.nlargest(30)

print(type(high_volume_publishers))
print(high_volume_publishers)

# Plot as a bar chart
plt.figure(figsize=(20, 6))
high_volume_publishers.plot(kind='bar', color='skyblue', edgecolor='black')

# Customize the chart
plt.xlabel("Publisher", fontsize=12)
plt.ylabel("Menge der Verffentlichungen", fontsize=12)
plt.title("Anzahl Verffentlichungen pro Publisher", fontsize=14)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
#Da Springer und Elsevier BV sehr gro sind, werden sie ber die 40000 Klausel einmal rausgenommen.
#Anschlieend werden die Top 30 genommen. Beim Verusch alle darzustellen ergibt sich eine unlesbare schwarze Wolke. 
lower_volume_publishers = counts_per_publisher[counts_per_publisher <= 40000].nlargest(30)
# Plot as a bar chart
plt.figure(figsize=(20, 6))
lower_volume_publishers.plot(kind='bar', color='skyblue', edgecolor='black')

# Customize the chart
plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Articles", fontsize=12)
plt.title("Number of Articles per Country", fontsize=14)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
