```python
#Forschungsaspekt: In welchen Publikationstypen wird publiziert?
import pandas as pd
import matplotlib.pyplot as plt
import glob

csv_folder_file_path = r"C:\HLAG\Apps\Self_Learn_Workspace\Python_Natural_Language_processing_NAKPrep\PY Hausarbeit\HA_TextAnalytics\lens_org_data_files\*.csv"
all_csv_file_paths = glob.glob(csv_folder_file_path)

individual_dataframes = [pd.read_csv(file) for file in all_csv_file_paths]
summarized_dataframe = pd.concat(individual_dataframes, ignore_index=True) 
```

```python
#Forschungsfrage: Wie hat sich das Volumen wissenschaftlicher Artikel ber Solarenergie in den letzten 10 Jahre entwickelt? 
columnName = ("Publication Type")
# Gruppierte Publikationsarten zhlen und sortieren
publicationTypeCounts = summarized_dataframe[columnName].dropna().astype(str).value_counts()

# Balkendiagramm erstellen
plt.figure(figsize=(12, 6))
publicationTypeCounts.plot(kind='barh', color='royalblue', edgecolor='black')

# Diagramm anpassen
plt.xlabel("Anzahl der Publikationen", fontsize=12)
plt.ylabel("Publikationstyp", fontsize=12)
plt.title("Anzahl wissenschaftlicher Publikationen zu Solarenergie pro Publikationstyp", fontsize=14)
plt.grid(axis='x', linestyle="--", alpha=0.7)

# Zeige den Plot
plt.show()
```

```python
publiations_tpes_other_than_journals = publicationTypeCounts[publicationTypeCounts <= 50000]

# Balkendiagramm erstellen
plt.figure(figsize=(12, 6))
publiations_tpes_other_than_journals.plot(kind='barh', color='royalblue', edgecolor='black')

# Diagramm anpassen
plt.xlabel("Anzahl der Publikationen", fontsize=12)
plt.ylabel("Publikationstyp", fontsize=12)
plt.title("Anzahl wissenschaftlicher Publikationen zu Solarenergie pro Publikationstyp ohne Journals", fontsize=14)
plt.grid(axis='x', linestyle="--", alpha=0.7)

# Zeige den Plot
plt.show()
```
