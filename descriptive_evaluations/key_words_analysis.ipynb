```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
from wordcloud import WordCloud, STOPWORDS

csv_folder_file_path = r"C:\HLAG\Apps\Self_Learn_Workspace\Python_Natural_Language_processing_NAKPrep\PY Hausarbeit\HA_TextAnalytics\lens_org_data_files\*.csv"
all_csv_file_paths = glob.glob(csv_folder_file_path)

individual_dataframes = [pd.read_csv(file) for file in all_csv_file_paths]
print(len(individual_dataframes))
summarized_dataframe = pd.concat(individual_dataframes, ignore_index=True)

columnName = ("Source Country")

country_counts = summarized_dataframe[columnName].dropna().astype(str).value_counts()
```

```python
#Top keywords (as word cloud) for the "Title"
columnName = "Title"
text = text = " ".join(summarized_dataframe[columnName].dropna().astype(str))
text = text.lower()
print(type(text))

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
# "How often which words are used in the title (removing big/small letters)"
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.show()
```

```python
#Top keywords (as word cloud) for the "Abstract"
import re

columnName = "Abstract"
text = text = " ".join(summarized_dataframe[columnName].dropna().astype(str))
text = text.lower()

#Data cleaning - kill the <jats> tags in the abstracts
text = re.sub(r"<.*?jats.*?>", "", text)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
# "How often which words are used in the title (removing big/small letters)"
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.show()
```
